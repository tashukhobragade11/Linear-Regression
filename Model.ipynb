```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv')
df.head()

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.scatter(df['Experience Years'],df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.show()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = df[['Experience Years']]
y = df['Salary']
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[10]])
```

```python
Year=float(input("Enter your Experience in year:"))
salary=lr.predict([[Year]])
print(salary)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.scatter(df['Experience Years'],df['Salary'])
plt.plot(x,lr.predict(x),color='red')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.show()
```
